```python
import numpy as np

# buat data jumlah kamar
bedrooms = np.array([1, 1, 2, 3, 3, 4, 4, 5, 5, 5])

# data harga rumah, kita asumsikan itu dalam dolar
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])
```

```python
import matplotlib.pyplot as plt   

%matplotlib inline

plt.scatter(bedrooms, house_price)
```

```python
from sklearn.linear_model import LinearRegression

# latih model dengan linearRegression.fit()

bedrooms = bedrooms.reshape(-1, 1)
linier = LinearRegression()
linier.fit(bedrooms, house_price)

# menampilkan plot hubungan antara jumlah kamar dan harga rumah
plt.scatter(bedrooms, house_price)
plt.plot(bedrooms, linier.predict(bedrooms))
```
